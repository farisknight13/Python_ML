```
import pandas as pd
from sklearn.metrics import mutual_info_score
import numpy as np

link = 'https://raw.githubusercontent.com/farisknight13/Python_ML/main/ObesityDataSet_raw_and_data_sinthetic.csv'
df = pd.read_csv(link)
df
```

# Drop Colume is String

```
df = df = df.drop(['Gender'],axis=1)
df = df = df.drop(['family_history_with_overweight'],axis=1)
df = df = df.drop(['FAVC'],axis=1)
df = df = df.drop(['CAEC'],axis=1)
df = df = df.drop(['SMOKE'],axis=1)
df = df = df.drop(['SCC'],axis=1)
df = df = df.drop(['CALC'],axis=1)
df = df = df.drop(['MTRANS'],axis=1)
df = df = df.drop(['NObeyesdad'],axis=1)
df
```

# Corrilation

```
corr_matrix = df.corr().abs()
corr_matrix
```

```
df['BMI'] = df['Weight']/df['Height']**2
df
```

```
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```
regr_nor = linear_model.LinearRegression()
regr_nor.fit(X_train,y_train)
pred_nor = regr_nor.predict(X_test)
```

```
test_nor = pd.DataFrame()
test_nor['ACTUAL'] = y_test
test_nor['Predict'] = pred_nor
test_nor['MAE%'] = (test_nor['ACTUAL']-test_nor['Predict']).abs()/test_nor['ACTUAL']*100
test_nor
```

```
print('intercept',regr_nor.intercept_)
```

```
print('coefficient',regr_nor.coef_)
```

```
print('adjusted r square',r2_score(y_test,pred_nor))
```
